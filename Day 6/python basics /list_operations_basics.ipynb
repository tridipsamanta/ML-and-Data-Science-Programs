## Write a program to take n elements from the user and store them in a list.

```python
n = int(input("Enter number of elements : "))
num = []
for i in range(n):
    element = input("Enter elements : ")
    num.append(element)
print('Final List : ',num)
```

## Write a program to find the sum of all elements in a list.

```python
list = [10,20,30,40,50]
print('Sum of all element in list is : ',sum(list))

#using loops manually.....
sum = 0
for i in list:
    sum += i
print('Using loop : ',sum)
```

## Write a program to find the largest element in a list.

```python
list = [12,34,67,2,-7]
print('Largest number/element in my list is : ',max(list))

#using loop manually....
larg = list[0]
for i in list:
    if i > larg:
        larg = i
print('Largest number is : ',larg)

```

## Write a program to find the smallest element in a list.

```python
list = [12,34,67,2,-7]
print('smallest number/element in my list is : ',min(list))

#using loop manually....
small = list[0]
for i in list:
    if i < small:
        larg = i
print('smallest number is : ',larg)
```

## Write a program to count even and odd numbers in a list.

```python
list = [1,2,3,4,5,6,7,8,9]
count =0
count1 = 0
for i in list:
    if i%2==0:
        count += 1
    elif i%2 != 0:
        count1 += 1
print('Even element : ',count,'Odd element : ',count1)
    
```
